# Bernoulli Process
We simulate the number of successes or arrivals in a Bernoulli process:

$$
X_n =
\begin{cases}
0 & \text{if } n = 0, \\
Y_1 + Y_2 + \cdots + Y_n & \text{if } n = 1,2,\dots
\end{cases}
$$

where $Y_1, Y_2, \dots, Y_n$ are independent Bernoulli random variables, all with the same parameter $p$:

$$
Y_i =
\begin{cases}
1 & \text{prob } p, \\
0 & \text{prob } 1-p \\
\end{cases}
$$

The expectation and variance of $X_n$ are given by:

$$
\mathbb{E}[X_n] = np \qquad \mathrm{Var}(X_n) = np(1-p)
$$

```python
N=100
num_proc = 10
domain = np.arange(start=0,stop=N+1,step=1)

p=0.5
plt.figure(figsize=(12,6))
for k in range(num_proc):
  individual_proc = BernoulliProcess(N=N,p=p, seed=k+1, a=1664525, m=2**32)
  plt.plot(domain, individual_proc,linewidth =0.7)

av = domain*p
st_dev = np.sqrt(domain*p*(1-p))
plt.plot(domain, av,color='black',label = 'Average')
plt.plot(domain,av+st_dev,color = 'red',label='Standard deviation')
plt.plot(domain, av-st_dev,color = 'red')

plt.xlabel('n',fontsize=20)
plt.ylabel('$X_n$',fontsize=20)

plt.title('Bernoulli Process Simulations')
plt.legend()
plt.show()
```
